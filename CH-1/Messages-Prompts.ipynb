# **PROMPTS**


```python
import os 
import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv

load_dotenv()

llm_openai = ChatOpenAI(model="gpt-4.1-mini", temperature=0)

user_input = input("Enter a topic : ")

dynamic_prompt = PromptTemplate.from_template("Write a fun fact about {topic}")

ready_prompt = dynamic_prompt.invoke({"topic" : user_input})

llm_openai.invoke(ready_prompt).content


```

# **TONE PROMPT**

```python
import os 
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
import load_dotenv
from dotenv import load_dotenv

load_dotenv()

llm_openai = ChatOpenAI(model="gpt-4.1-mini", temperature=0)

user_input = input("Enter a topic : ")

prompt_template = ChatPromptTemplate.from_messages([
  ("system", "You are an automobile enthusiast"),
  ("user", "Write a interesting fact about {topic}")
])

ready_prompt = prompt_template.invoke({"topic" : user_input})

llm_openai.invoke(ready_prompt).content
```
