```python
import os 
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from dotenv import load_dotenv
from pydantic import BaseModel, Field
from typing import TypedDict

load_dotenv()

llm_openai = ChatOpenAI(model="gpt-4.1-mini", temperature=0)
```

# **Guiding in Prompts**

```python
result = llm_openai.invoke("Tell me joke. Generate the output in key-value pair where key is the setup and value is the punchline")

result.content
```

# **Using Pydantic**

```python
class llm_schema(BaseModel):
  setup : str = Field(description = "The setup of the joke ")
  punchline : str = Field(description = "The punchline of the joke")

llm_structured_output = llm_openai.with_structured_output(llm_schema)

result = llm_structured_output.invoke("Tell me joke about cars")

result
```

# **Using TypedDict**

```python
class llm_schema_td(TypedDict):
  setup : str 
  punchline : str

llm_structured_output = llm_openai.with_structured_output(llm_schema_td)

result = llm_structured_output.invoke("Tell me a joke about Ferrari")

result
```
