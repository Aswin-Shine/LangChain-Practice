```python
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda, RunnableParallel
from dotenv import load_dotenv

load_dotenv()

llm_openai = ChatOpenAI(model="gpt-4.1-mini", temperature=0)

str_parser = StrOutputParser()
```

```python
# Task 1 - Prompt
prompt_template = ChatPromptTemplate.from_messages([
  ("system", "You are a Formula 1 Grand Prix Summarizer"),
  ("user", "Breif summary of the race : {input}")
])

# Task 2 - LLM

# Task 3 - Str Parser

# Task 4 - Make Dictionary 
def dictionary_maker(text:str)-> dict :
  return {"text" : text}
dictionary_maker_runnable = RunnableLambda(dictionary_maker)
```

# **Parallel Chain 1**

```python
def linkedin_chain(text:dict):
  text = text["text"]

  # Task 1 - Prompt
  linkedin_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an automotive linkedin post generator"),
    ("user", "Create a linkedin post from the following : {text}")
  ])

  # Task 2 - LLM

  # Task 3 - Str Parser

  chain_linkedin = linkedin_prompt | llm_openai | str_parser
  result = chain_linkedin.invoke(text)
  return result

chain_linkedin_runnable = RunnableLambda(linkedin_chain)


```

# **Parallel Chain 2**


```python
def insta_chain(text:dict):
  text = text["text"]

  # Task 1 - Prompt 
  insta_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an automotive instagram post generator"),
    ("user", "Create a instagram post from the following : {text}")
  ])

  # Task 2 - LLM

  # Task 3 - Str Parser

  chain_insta = insta_prompt | llm_openai | str_parser
  result = chain_insta.invoke(text)
  return result

chain_insta_runnable = RunnableLambda(insta_chain)
```

# **Final Orchestration**

```python
final_chain = prompt_template | llm_openai | str_parser | dictionary_maker_runnable | RunnableParallel(branches = {"linkedin" : chain_linkedin_runnable, "instagram" : chain_insta_runnable})

final_chain.invoke("2024 British Grand Prix ")
```

# **Beautified Output**

```python
def beautify(final_response:dict)-> dict:
  linkedin_response = final_response['branches']['linkedin']
  instagram_response = final_response['branches']['instagram']

  return {"linked" : linkedin_response, "instagram" : instagram_response}

beautify_runnable = RunnableLambda(beautify)

beautified_chain = final_chain | beautify_runnable
beautified_chain.invoke("1984 Monaco Grand Prix")
```
